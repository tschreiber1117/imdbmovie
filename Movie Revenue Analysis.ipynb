## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python

top5 = df.groupby('production_companies')['original_title', 'release_date'].count().reset_index().sort_values('original_title', ascending=False)
top5

```

### What 5 movie genres grossed the highest all-time?

```python

hgenre = df.groupby('genres')['revenue'].mean().reset_index().sort_values('revenue', ascending=False)
hgenre
```

### Who are the top 5 grossing directors?

```python
hdirector = df.groupby('director')['revenue'].sum().reset_index().sort_values('revenue', ascending=False)
hdirector
```

### Compare the revenue of the highest grossing movies of all time.

```python
hmovie = df.groupby('original_title')['revenue'].sum().reset_index().sort_values('revenue', ascending=False)
hmovie
```
